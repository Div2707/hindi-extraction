```python
from PIL import Image
import pytesseract

# Set the path for Tesseract-OCR executable
pytesseract.pytesseract.tesseract_cmd = r'<path_to_tesseract>'

def extract_text_from_pdf(pdf_path):
    # Open the PDF using PIL
    pdf = Image.open(pdf_path)

    # Initialize a variable to store extracted text
    extracted_text = ""

    # Loop through each page in the PDF
    for page in range(pdf.n_frames):
        pdf.seek(page)
        page_img = pdf.convert("RGB")
        text = pytesseract.image_to_string(page_img, lang='hin') # 'hin' indicates Hindi language

        # Append the extracted text to the variable
        extracted_text += text

    return extracted_text

# Usage
pdf_path = 'EROLLGEN-S20-62-FinalRoll-Revision1-HIN-1-WI_page-0002.jpg'
extracted_text = extract_text_from_pdf(pdf_path)

# Print or save the extracted text
print(extracted_text)
```
